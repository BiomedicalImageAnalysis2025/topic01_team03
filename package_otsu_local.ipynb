# Otsu local 

```python
import os
import numpy as np
import matplotlib.pyplot as plt

from skimage import io, img_as_ubyte
from skimage.filters import rank
from skimage.morphology import disk
from src.image_find import load_image_gray

%pip install imagecodecs

# 2) Bild laden (als Float [0,1]) und in uint8 (0255) umwandeln:
image =  load_image_gray("dna-44.png")
image_u8   = img_as_ubyte(image)



# 3) Lokalen OtsuFilter aus skimage.filters.rank importieren
#    Dafr brauchst du 8-Bit-Bild UND einen strukturellen Element (footprint) Kreis.
radius = 50  # halbe Fenstergre ~ 7  1515 Pixel-Fenster
footprint = disk(radius)

# 4) rank.otsu berechnet fr **jeden Pixel** den Otsu-Schwellwert
#    innerhalb des kreisfrmigen Nachbarschaftsbereichs (footprint).
t_local_rank = rank.otsu(image_u8, footprint)

# 5) Erzeuge die Binrmaske: Werte > lokaler Threshold  1; sonst 0
binary_local_rank = image_u8 > t_local_rank

# 6) Anzeigen
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].imshow(image_u8, cmap="gray", vmin=0, vmax=255, interpolation="none")
axes[0].set_title("Original (uint8)") 
axes[0].axis("off")

axes[1].imshow(binary_local_rank, cmap="gray", interpolation="none")
axes[1].set_title("Binrmaske (lokaler Otsu)")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
